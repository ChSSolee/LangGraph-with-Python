## 01. 랭그래프 (LangGraph) 소개
- https://www.langchain.com/langgraph
- LangGraph는 자연어 처리와 AI 응용 프로그램 개발을 위한 강력한 프레임워크로, 복잡한 언어 모델과의 상호작용을 효율적이고 구조화된 방식으로 구현할 수 있도록 도움. 
- 다양한 데이터 소스와 언어 모델을 통합하여 지능형 응답 생성, 정보 검색, 텍스트 분석 등의 고도화된 시스템을 구축할 수 있음

<br>

### LangGraph 첫걸음
- **LangGraph는 단계별로 실행되는 작업들을 연결하는 도구**
- **각각의 작업(노드)을 화살표(엣지)로 연결해서 하나의 완성된 프로그램을 개발**

<br>

#### 간단한 구현

````python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
````

1. **데이터 저장소 만들기 (State)**

````python
class MyState(TypedDict):
    message: str
````

2. **작업 함수 만들기 (Node)**

````python
def say_hello(state):
    return {"message": "Hello, LangGraph!"}
````

3. **그래프 만들기**

````python
graph = StateGraph(MyState)
graph.add_node("hello", say_hello)
graph.add_edge(START, "hello")
graph.add_edge("hello", END)
````

4. **실행하기**

````python
app = graph.compile()
result = app.invoke({"message": ""})
print(result)
````

#### 주요 목적:
- **LangGraph는 복잡한 워크플로우와 정교한 의사결정 프로세스를 구현하는 데 특화** $\rightarrow$ 여러 단계의 처리 과정이나 조건부 로직이 필요한 고급 AI 시스템에 적합
- LangChain은 LLM을 다양한 외부 도구와 쉽게 통합하고, 간단한 체인 구조로 애플리케이션을 구성하는 데 중점 $\rightarrow$ **빠른 프로토타이핑과 기본적인 LLM 기반 애플리케이션 개발에 유용**

<br>

#### 구조:
- LangGraph는 그래프 기반 구조를 채택하여 노드와 엣지로 구성된 **유연한 워크플로우**를 생성 $\rightarrow$ 복잡한 로직과 다단계 프로세스를 직관적으로 모델링하는 데 도움
- LangChain은 체인과 에이전트 기반 구조를 사용 $\rightarrow$ **선형적인 처리 과정이나 미리 정의된 에이전트 패턴을 구현하기 쉽게 만듬**

<br>

#### 상태 관리:
- LangGraph는 **명시적이고 세밀한 상태 관리**를 제공. 개발자가 각 단계에서 상태를 직접 제어하고 수정할 수 있어, 복잡한 상태 변화를 정확하게 추적하고 관리할 수 있음
- LangChain은 **상대적으로 암시적이고 자동화된 상태 관리를 제공** $\rightarrow$ 이는 개발 과정을 단순화하지만, 세부적인 상태 제어가 필요한 경우 제한적일 수 있습니다.

<br>

#### 유연성:
- LangGraph는 높은 유연성을 제공하여 커스텀 로직을 쉽게 구현. **개발자가 원하는 대로 그래프 구조를 설계하고 각 노드의 동작을 상세하게 정의**
- LangChain은 미리 정의된 컴포넌트를 중심으로 구성되어 있어, 기본적인 기능을 빠르게 구현할 수 있지만 **고도로 커스터마이즈된 로직을 구현하는 데는 상대적으로 제한**

<br>

#### 학습 곡선:
- LangGraph는 그래프 이론과 상태 관리에 대한 이해가 필요하여 상대적으로 가파른 러닝 커브. 이를 통해 더 복잡하고 정교한 시스템을 구축
- LangChain은 직관적인 API와 풍부한 예제로 인해 상대적으로 완만한 러닝 커브. 초보자도 빠르게 기본적인 LLM 애플리케이션을 만들 수 있음.

<br>

#### 용도:
- **LangGraph는 복잡한 AI 시스템이나 다중 에이전트 시스템을 구축하는 데 적합** $\rightarrow$  여러 AI 모델이 상호작용하거나 복잡한 의사결정 과정이 필요한 프로젝트에 이상적
- LangChain은 간단한 LLM 애플리케이션이나 RAG(Retrieval-Augmented Generation) 시스템을 구축하는 데 주로 사용 $\rightarrow$ 빠른 개발과 프로토타이핑이 필요한 프로젝트에 적합

<br>

> **복잡하고 맞춤화된 AI 시스템이 필요하다면 LangGraph가, 빠른 개발과 간단한 LLM 통합이 목표라면 LangChain이 더 적합**

<br>



<br>

### LangGraph의 효용
- **복잡한 로직 구현**: 다단계 의사결정 프로세스나 복잡한 워크플로우를 쉽게 구현
- **세밀한 제어**: 애플리케이션의 각 단계를 정밀하게 제어할 수 있어, 고도로 커스터마이즈된 동작을 구현
- **확장성**: 서브그래프를 통해 대규모 시스템을 모듈화하여 관리
- **상태 지속성**: 체크포인팅을 통해 장기 실행 태스크와 오류 복구를 쉽게 구현
- **다중 에이전트 시스템**: 여러 AI 에이전트의 상호작용을 효과적으로 모델링

<br>



### 기본 모델 설정

````python
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

load_dotenv()
````

<br>

#### `init_chat_model`을 활용한 LLM 설정 (권장)

````python
gpt4_mini = init_chat_model(
    "openai:gpt-4.1-mini",
    temperature=0.7,
    max_tokens=150,
)

gpt4 = init_chat_model(
    "openai:gpt-4.1",
    temperature=0.7,
    max_tokens=300,
)
````

#### `ChatOpenAI`를 직접 사용

````python
from langchain_openai import ChatOpenAI
````

````python
gpt4_mini = ChatOpenAI(
    model="gpt-4.1-mini",
    temperature=0.7,
    max_tokens=150,
)

gpt4 = ChatOpenAI(
    model="gpt-4.1",
    temperature=0.7,
    max_tokens=300,
)
````

#### 모델 사용 예시

````python
from langchain_core.messages import HumanMessage
````

````python
# GPT-4.1-mini 사용
response_mini = gpt4_mini.invoke([HumanMessage(content="Hello, how are you?")])
print(response_mini.content)

# GPT-4.1 사용
response_full = gpt4.invoke([HumanMessage(content="Explain the concept of machine learning.")])
print(response_full.content)
````

### 다양한 LLM 모델 활용 방법 (Anthropic Claude, Google Gemini)

```python
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

# .env 파일에서 환경 변수 로드
load_dotenv()

# Anthropic Claude 설정
claude = init_chat_model("anthropic:claude-sonnet-4-6")

# Google Gemini 설정
gemini = init_chat_model("google-genai:gemini-2.5-flash")

# OpenAI GPT (비교를 위해)
gpt = init_chat_model("openai:gpt-4.1-mini")

# 동일한 인터페이스로 사용
claude_response = claude.invoke("Explain the concept of Directed Graph.")
print("Claude's response:", claude_response.content)

gemini_response = gemini.invoke("Explain the components of LangGraph.")
print("Gemini's response:", gemini_response.content)
```
